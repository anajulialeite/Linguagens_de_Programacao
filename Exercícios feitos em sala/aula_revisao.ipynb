# Aula de Revisão – Linguagem Python

## 1. Tipagem em Python

Python é uma linguagem de **tipagem dinâmica**: não é necessário
declarar explicitamente o tipo da variável.  
Entretanto, é **fortemente tipada**, pois não permite operações
inválidas entre tipos sem conversão.

``` python
x = 10        # inteiro
x = "dez"     # string (mesma variável, outro tipo)

# Operação inválida
# print(10 + "5")  # Gera erro

# Conversão correta
print(10 + int("5"))   # 15
print("Resultado: " + str(10))  # Resultado: 10
```

## 2. Segurança no Armazenamento de Dados

Armazenar senhas em texto puro é uma prática insegura. O correto é
utilizar funções de **hash criptográfico**.

``` python
import hashlib

senha = "minha_senha_secreta"
hash_senha = hashlib.sha256(senha.encode()).hexdigest()
print("Senha criptografada:", hash_senha)

# Comparação de autenticação
entrada = "minha_senha_secreta"
if hashlib.sha256(entrada.encode()).hexdigest() == hash_senha:
    print("Acesso permitido")
else:
    print("Acesso negado")
```

## 3. Estruturas de Repetição

### Exemplo com `for`:

``` python
for i in range(1, 6):
    print(i)
```

### Exemplo com `while`:

``` python
contador = 0
while contador < 5:
    print(contador)
    contador += 1
```

### Controle de fluxo:

``` python
for i in range(10):
    if i == 3:
        continue  # Pula para a próxima iteração
    if i == 7:
        break     # Interrompe o laço
    print(i)
```

## 4. Função `range()`

``` python
print(list(range(5)))        # [0, 1, 2, 3, 4]
print(list(range(2, 6)))     # [2, 3, 4, 5]
print(list(range(0, 10, 2))) # [0, 2, 4, 6, 8]
```

## 6. Loops Infinitos

## 7. Encapsulamento (POO)

``` python
class Conta:
    def __init__(self, saldo):
        self._saldo = saldo  # Atributo protegido
    
    @property
    def saldo(self):
        return self._saldo
    
    def depositar(self, valor):
        if valor > 0:
            self._saldo += valor

c = Conta(100)
c.depositar(50)
print(c.saldo)  # 150
```

## 8. Herança e Polimorfismo (POO)

``` python
class Usuario:
    def __init__(self, nome):
        self.nome = nome
    
    def acessar(self):
        return "Acesso básico"

class Admin(Usuario):
    def acessar(self):
        return "Acesso administrativo"

class Visitante(Usuario):
    def acessar(self):
        return "Acesso limitado"

usuarios = [Usuario("João"), Admin("Maria"), Visitante("Pedro")]
for u in usuarios:
    print(u.nome, "-", u.acessar())
```
